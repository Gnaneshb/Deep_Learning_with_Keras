```python
#path to the directory where the original dataset was compressed
original_dataset_dir = 'C:\\Users\\Ros\\Desktop\\Gnanesh\\Deep Learning with Keras\\Dogs_vs_cats\\train'
```

```python
#directory to store small dataset
base_dir = 'C:\\Users\\Ros\\Desktop\\Gnanesh\\Deep Learning with Keras\\dogs_vs_cats_small'
os.mkdir(base_dir)
```

```python
#directories for the trainging, validation and test splits
train_dir = os.path.join(base_dir,'train')
os.mkdir(train_dir)
```

```python
test_dir = os.path.join(base_dir,'test')
os.mkdir(test_dir)
```

```python
validation_dir = os.path.join(base_dir,'validation')
os.mkdir(validation_dir)
```

```python
#directory with cat training images
train_cat_dir = os.path.join(train_dir,'cats')
os.mkdir(train_cat_dir)
```

```python
#directory with dog training images
train_dog_dir = os.path.join(train_dir,'dogs')
os.mkdir(train_dog_dir)
```

```python
#directory with cat testing images
test_cat_dir = os.path.join(test_dir,'cats')
os.mkdir(test_cat_dir)
```

```python
#directory with dog training images
test_dog_dir = os.path.join(test_dir,'dogs')
os.mkdir(test_dog_dir)
```

```python
#directory with dog validation images
validation_dog_dir = os.path.join(validation_dir,'dogs')
os.mkdir(validation_dog_dir)
```

```python
#directory with cat validation images
validation_cat_dir = os.path.join(validation_dir,'cats')
os.mkdir(validation_cat_dir)
```

```python
#copies the first 1000 cat images to train_cat_dir
fnames = ['cat.{}.jpg'.format(i) for i in range(1000)]
```

```python
fname
```

```python
for fname in fnames:
    src = os.path.join(original_dataset_dir, fname)
    dst = os.path.join(train_cat_dir,fname)
    shutil.copyfile(src, dst)
```

```python
#copies next 500 cat images to test_cat_dir
fnames = ['cat.{}.jpg'.format(i) for i in range(1000,1500)]

for fname in fnames:
    src = os.path.join(original_dataset_dir, fname)
    dst = os.path.join(test_cat_dir,fname)
    shutil.copyfile(src, dst)
```

```python
#copies next 500 cat images to validation_cat_dir
fnames = ['cat.{}.jpg'.format(i) for i in range(1500,2000)]

for fname in fnames:
    src = os.path.join(original_dataset_dir, fname)
    dst = os.path.join(validation_cat_dir,fname)
    shutil.copyfile(src, dst)
```

```python
#copies next 500 dog images to test_dog_dir
fnames = ['dog.{}.jpg'.format(i) for i in range(1000,1500)]

for fname in fnames:
    src = os.path.join(original_dataset_dir, fname)
    dst = os.path.join(test_dog_dir,fname)
    shutil.copyfile(src, dst)
```
